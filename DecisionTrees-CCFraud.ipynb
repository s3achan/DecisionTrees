Import the required files

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

Load Data & Keep Only 4 Features

```python


df = pd.read_csv("creditcard.csv")

features = ['V14', 'V10', 'V12', 'Amount']
X = df[features]
y = df['Class']

X.head()
```

Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```
